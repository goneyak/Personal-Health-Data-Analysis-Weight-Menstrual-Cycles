```python
from scipy.stats import ttest_rel
import numpy as np

def test_weight_change(event_dates, df_weight, before_days=3, after_days=3):
    before_weights = []
    after_weights = []
    
    for date in event_dates:
        if pd.isna(date):
            continue
        
        before_range = pd.date_range(date - pd.Timedelta(days=before_days), date - pd.Timedelta(days=1))
        after_range = pd.date_range(date + pd.Timedelta(days=1), date + pd.Timedelta(days=after_days))
        
        before_vals = df_weight.loc[df_weight['Date'].isin(before_range), 'Weight']
        after_vals = df_weight.loc[df_weight['Date'].isin(after_range), 'Weight']
        
        if not before_vals.empty and not after_vals.empty:
            before_weights.append(before_vals.mean())
            after_weights.append(after_vals.mean())
    
    before_weights = np.array(before_weights)
    after_weights = np.array(after_weights)
    
    # Paired t-test
    t_stat, p_value = ttest_rel(before_weights, after_weights)
    mean_diff = after_weights.mean() - before_weights.mean()
    
    return {
        'before_mean': before_weights.mean(),
        'after_mean': after_weights.mean(),
        'mean_diff': mean_diff,
        't_stat': t_stat,
        'p_value': p_value
    }

# 생리 시작일 기준
period_start_stats = test_weight_change(df_periods['begin'], df_weight, before_days=3, after_days=3)

# 배란일 기준
ovulation_stats = test_weight_change(df_periods['ovulation_date'], df_weight, before_days=3, after_days=3)

print("=== Period Start ===")
print(period_start_stats)

print("\n=== Ovulation ===")
print(ovulation_stats)
```

# 1) Menstrual Cycle Alignment Analysis with Weight Data

## 📌 Objective
The goal of this analysis is to investigate whether my body weight shows a consistent and significant pattern in relation to my menstrual cycle.

Specifically, we want to know:
1. Does my weight tend to increase or decrease before and after **period start**?
2. Does my weight show a consistent trend around **ovulation**?
3. Are there distinct phases within the cycle where weight is noticeably higher or lower?

## 🔍 Analysis Approach

### Step 1 — Data Preparation
- **Weight data**: Daily weight measurements (missing days interpolated).
- **Cycle data**: Menstrual period start dates, ovulation dates, and cycle lengths.

Both datasets are merged or used in parallel, with a shared date reference.

```python
import numpy as np
import pandas as pd

# ===== Step 1: Function to align weight data around specific event dates =====
def align_weight(df_weight, event_dates, window):
    """
    Aligns daily weight data around given event dates (e.g., period start, ovulation).
    
    Parameters:
        df_weight (pd.DataFrame): DataFrame with columns ['Date', 'Weight'].
        event_dates (iterable): List/Series of event dates (datetime).
        window (int): Number of days before and after the event date to include.
        
    Returns:
        np.ndarray: 2D array where each row is one event cycle,
                    and columns are days from -window to +window.
    """
    # Create the range of days relative to the event (e.g., -14 ~ +14)
    day_range = np.arange(-window, window + 1)
    aligned_cycles = []

    for start_date in event_dates:
        if pd.isna(start_date):
            continue  # Skip if the event date is missing

        # Define the date range around the event
        start_window = start_date - pd.Timedelta(days=window)
        end_window = start_date + pd.Timedelta(days=window)

        # Extract weight data within the window
        df_window = df_weight[(df_weight['Date'] >= start_window) &
                              (df_weight['Date'] <= end_window)].copy()

        # Calculate relative day in cycle (0 = event date)
        df_window['cycle_day'] = (df_window['Date'] - start_date).dt.days

        # Reindex to ensure the same number of days for all cycles
        df_window = df_window.set_index('cycle_day').reindex(day_range)

        # Store the aligned weight values (NaN for missing days)
        aligned_cycles.append(df_window['Weight'].values)

    return np.array(aligned_cycles, dtype=float)

# ===== Step 2: Apply the function for both Period Start and Ovulation =====
window = 7  # +/- 14 days

# Align to period start dates
period_start_dates = df_periods['begin']
aligned_period_start = align_weight(df_weight, period_start_dates, window)

# Align to ovulation dates
ovulation_dates = df_periods['ovulation_date']
aligned_ovulation = align_weight(df_weight, ovulation_dates, window)

# Print shape for verification
print("Period Start aligned shape:", aligned_period_start.shape)
print("Ovulation aligned shape:", aligned_ovulation.shape)
```

### Step 2 — Event-based Alignment
We align weight data around two key events:
- **Period start** (Day 0)
- **Ovulation date** (Day 0)

For each event date:
- Extract a fixed time window (e.g., **±14 days**).
- Normalize each cycle to a relative day index (`-14 ~ +14`).
- Store all cycles in an aligned 2D matrix for averaging.

```python
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt
import numpy as np

def analyze_event_pattern(aligned_data, window, event_name):
    """
    Analyze weight pattern around a given event.
    
    Parameters:
        aligned_data (np.ndarray): 2D array, each row is one cycle, columns are days (-window ~ +window)
        window (int): Number of days before/after event
        event_name (str): Name of the event (for printing and plotting)
    """
    day_range = np.arange(-window, window + 1)
    
    # Calculate the mean weight across all cycles (ignoring NaN)
    mean_cycle = np.nanmean(aligned_data, axis=0)
    
    # Calculate average weight before and after event
    before_mean = np.nanmean(aligned_data[:, :window])   # days < 0
    after_mean = np.nanmean(aligned_data[:, window+1:])  # days > 0
    mean_diff = after_mean - before_mean
    
    # Perform paired t-test (only rows without NaN in before/after)
    valid_rows = ~np.isnan(np.nanmean(aligned_data, axis=1))
    t_stat, p_value = ttest_rel(
        np.nanmean(aligned_data[valid_rows, :window], axis=1),
        np.nanmean(aligned_data[valid_rows, window+1:], axis=1)
    )
    
    # Print results
    print(f"=== {event_name} ===")
    print({
        "before_mean": before_mean,
        "after_mean": after_mean,
        "mean_diff": mean_diff,
        "t_stat": t_stat,
        "p_value": p_value
    })
    
    # Visualization
    plt.figure(figsize=(12, 6))
    
    # Plot individual cycles (gray lines)
    for cycle in aligned_data:
        plt.plot(day_range, cycle, color='gray', alpha=0.3)
    
    # Plot average line (red)
    plt.plot(day_range, mean_cycle, color='red', linewidth=2, label='Average')
    
    # Vertical line at event day
    plt.axvline(0, color='black', linestyle='--', label=event_name)
    
    plt.xlabel('Days from Event')
    plt.ylabel('Weight (kg)')
    plt.title(f'Weight Pattern Around {event_name} (±{window} days)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Run for both events
analyze_event_pattern(aligned_period_start, window, "Period Start")
analyze_event_pattern(aligned_ovulation, window, "Ovulation")
```

### Step 3 — Pattern Analysis
From the aligned dataset:
- Compute **mean weight change** over the window.
- Compare **before vs after** averages (t-test or Wilcoxon test).
- Visualize patterns:
  - Multiple gray lines for individual cycles
  - A bold red line for the average trend
  - A vertical dashed line marking the event date

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_event_comparison_with_ci(aligned_data1, aligned_data2, window, label1, label2):
    """
    Plot average weight patterns for two events with 95% confidence intervals.
    """
    day_range = np.arange(-window, window + 1)
    
    def mean_ci(data):
        mean = np.nanmean(data, axis=0)
        std_err = np.nanstd(data, axis=0) / np.sqrt(np.sum(~np.isnan(data), axis=0))
        ci = 1.96 * std_err  # 95% CI
        return mean, ci
    
    # Event 1 stats
    mean1, ci1 = mean_ci(aligned_data1)
    # Event 2 stats
    mean2, ci2 = mean_ci(aligned_data2)
    
    plt.figure(figsize=(12, 6))
    
    # Plot Event 1 with CI shading
    plt.plot(day_range, mean1, color='red', linewidth=2, label=label1)
    plt.fill_between(day_range, mean1 - ci1, mean1 + ci1, color='red', alpha=0.2)
    plt.axvline(0, color='red', linestyle='--', alpha=0.5)
    
    # Plot Event 2 with CI shading
    plt.plot(day_range, mean2, color='blue', linewidth=2, label=label2)
    plt.fill_between(day_range, mean2 - ci2, mean2 + ci2, color='blue', alpha=0.2)
    plt.axvline(0, color='blue', linestyle='--', alpha=0.5)
    
    plt.xlabel('Days from Event')
    plt.ylabel('Weight (kg)')
    plt.title(f'Average Weight Pattern with 95% CI: {label1} vs {label2} (±{window} days)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Run the comparison plot with CI
plot_event_comparison_with_ci(
    aligned_period_start_7,
    aligned_ovulation_7,
    window=7,
    label1="Period Start",
    label2="Ovulation"
)
```

### Step 4 — Interpretation
- If the mean weight change is small and p-value is large → likely **no consistent pattern**.
- If there is a clear trend (e.g., weight increase before period start) → further phase-based analysis (follicular, ovulatory, luteal).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# =========================
# Step 0: Config
# =========================
WINDOW = 7                 # +/- days around the event (Period Start)
MIN_VALID_POINTS = 5       # minimum non-NaN points required in a cycle (before interpolation)
K_RANGE = range(2, 7)      # candidate number of clusters
RANDOM_STATE = 42
N_INIT = 10

# =========================
# Step 1: Align weight to Period Start (±WINDOW)
# =========================
def align_weight(df_weight, event_dates, window):
    """
    Align daily weight around event dates (Period Start).
    Returns a 2D array [n_cycles x (2*window+1)] with NaNs for missing days.
    """
    day_range = np.arange(-window, window + 1)
    aligned = []

    for start_date in event_dates:
        if pd.isna(start_date):
            continue

        start_win = start_date - pd.Timedelta(days=window)
        end_win = start_date + pd.Timedelta(days=window)

        w = df_weight[(df_weight['Date'] >= start_win) & (df_weight['Date'] <= end_win)].copy()
        w['cycle_day'] = (w['Date'] - start_date).dt.days

        # Reindex to full range so every cycle has same length
        w = w.set_index('cycle_day').reindex(day_range)

        aligned.append(w['Weight'].values)

    return np.array(aligned, dtype=float), day_range

aligned_raw, day_range = align_weight(df_weight, df_periods['begin'], WINDOW)

# Optional: keep mapping from row -> cycle start date for later inspection
cycle_starts = df_periods['begin'].dropna().reset_index(drop=True)

# =========================
# Step 2: Filter + Interpolate
# - Drop cycles with too few points
# - Linearly interpolate missing values within each cycle
# =========================
def interpolate_cycles(arr, min_valid=5):
    """
    Interpolate NaNs row-wise. Drops rows with too few non-NaN points before interpolation.
    Returns interpolated array and index mapping kept_rows (original indices).
    """
    kept_rows = []
    interp_rows = []
    for i, row in enumerate(arr):
        # Count valid points prior to interpolation
        valid = np.sum(~np.isnan(row))
        if valid < min_valid:
            continue
        # Interpolate along the fixed index day_range using pandas Series
        s = pd.Series(row, index=day_range, dtype=float)
        s = s.interpolate(method='linear', limit_direction='both')
        interp_rows.append(s.values)
        kept_rows.append(i)
    return np.array(interp_rows, dtype=float), np.array(kept_rows, dtype=int)

aligned_interp, kept_idx = interpolate_cycles(aligned_raw, min_valid=MIN_VALID_POINTS)
cycle_starts_kept = cycle_starts.iloc[kept_idx].reset_index(drop=True)

# =========================
# Step 3: Normalize per cycle (shape-only)
# - Z-score per row so clustering focuses on pattern, not absolute level
# =========================
def zscore_rows(arr):
    arr_mean = np.mean(arr, axis=1, keepdims=True)
    arr_std = np.std(arr, axis=1, keepdims=True)
    # Avoid division by zero: if std==0, keep zeros after centering
    arr_std[arr_std == 0] = 1.0
    return (arr - arr_mean) / arr_std

X = zscore_rows(aligned_interp)

# =========================
# Step 4: Choose K (number of clusters)
# - Use silhouette score over K_RANGE, pick the best
# =========================
sil_scores = []
best_k = None
best_score = -np.inf

for k in K_RANGE:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=N_INIT)
    labels = km.fit_predict(X)
    # Silhouette requires > 1 label and < n_samples labels
    if len(np.unique(labels)) > 1 and len(np.unique(labels)) < len(labels):
        score = silhouette_score(X, labels)
    else:
        score = -np.inf
    sil_scores.append((k, score))
    if score > best_score:
        best_score = score
        best_k = k

print("Silhouette scores (K, score):", sil_scores)
print("Chosen K (best by silhouette):", best_k)

# =========================
# Step 5: Fit final KMeans and get labels
# =========================
kmeans = KMeans(n_clusters=best_k, random_state=RANDOM_STATE, n_init=N_INIT)
labels = kmeans.fit_predict(X)

# Attach labels back to cycle meta
clustered_meta = pd.DataFrame({
    'cycle_start': cycle_starts_kept,
    'cluster': labels
}).sort_values('cycle_start').reset_index(drop=True)

print("\nCluster counts:")
print(clustered_meta['cluster'].value_counts().sort_index())

# =========================
# Step 6: Plot cluster mean patterns
# - One chart per cluster: thin lines for members + bold mean
# =========================
for c in range(best_k):
    idx = np.where(labels == c)[0]
    members = X[idx]  # z-scored (shape)
    mean_curve = np.mean(members, axis=0)

    plt.figure(figsize=(10, 5))
    # Plot members (thin)
    for row in members:
        plt.plot(day_range, row, alpha=0.3)

    # Plot mean (bold)
    plt.plot(day_range, mean_curve, linewidth=2, label=f'Cluster {c} mean')

    plt.axvline(0, linestyle='--', label='Period Start (Day 0)')
    plt.title(f'Cluster {c} — Aligned Weight Pattern (z-scored), n={len(idx)}')
    plt.xlabel('Days from Period Start')
    plt.ylabel('Z-scored Weight')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# =========================
# Step 7: (Optional) Inspect cluster membership table
# =========================
# You can merge back to original periods if needed:
# df_periods_joined = df_periods.merge(clustered_meta, left_on='begin', right_on='cycle_start', how='left')
# display(df_periods_joined.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===== Merge cluster labels back to period table =====
# Assumes df_periods includes: begin, end, duration_days, next_begin, cycle_length, ...
df_periods_joined = df_periods.merge(
    clustered_meta, left_on='begin', right_on='cycle_start', how='inner'
).copy()

# Derive Year / Season
def get_season(dt):
    m = dt.month
    if m in [12, 1, 2]:
        return 'Winter'
    elif m in [3, 4, 5]:
        return 'Spring'
    elif m in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df_periods_joined['year'] = df_periods_joined['begin'].dt.year
df_periods_joined['season'] = df_periods_joined['begin'].apply(get_season)

# ===== Cluster-level summary stats =====
summary = df_periods_joined.groupby('cluster').agg(
    n_cycles=('begin', 'count'),
    mean_cycle_length=('cycle_length', 'mean'),
    sd_cycle_length=('cycle_length', 'std'),
    mean_duration=('duration_days', 'mean'),
    sd_duration=('duration_days', 'std')
).round(2)

print("Cluster summary:")
print(summary)

# ===== Distributions by season / year =====
season_dist = pd.crosstab(df_periods_joined['cluster'], df_periods_joined['season']).sort_index(axis=1)
year_dist   = pd.crosstab(df_periods_joined['cluster'], df_periods_joined['year']).sort_index(axis=1)

print("\nSeason counts by cluster:")
print(season_dist)
print("\nYear counts by cluster:")
print(year_dist)

# Also percentage tables
season_pct = (season_dist.div(season_dist.sum(axis=1), axis=0) * 100).round(1)
year_pct   = (year_dist.div(year_dist.sum(axis=1), axis=0) * 100).round(1)

print("\nSeason % by cluster:")
print(season_pct)
print("\nYear % by cluster:")
print(year_pct)

# ===== Simple visuals (matplotlib only) =====

# 1) Mean cycle length & duration by cluster
fig, ax = plt.subplots(figsize=(8, 4))
x = np.arange(summary.shape[0])
width = 0.35
ax.bar(x - width/2, summary['mean_cycle_length'], width, label='Mean cycle length (days)')
ax.bar(x + width/2, summary['mean_duration'],     width, label='Mean period duration (days)')
ax.set_xticks(x)
ax.set_xticklabels([f'Cluster {c}' for c in summary.index])
ax.set_ylabel('Days')
ax.set_title('Mean cycle length & period duration by cluster')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 2) Season distribution (stacked %)
fig, ax = plt.subplots(figsize=(9, 4))
bottom = np.zeros(len(season_pct))
for col in season_pct.columns:  # Winter, Spring, Summer, Fall
    ax.bar(season_pct.index, season_pct[col], bottom=bottom, label=col)
    bottom += season_pct[col].values
ax.set_xticks(season_pct.index)
ax.set_xticklabels([f'Cluster {c}' for c in season_pct.index])
ax.set_ylabel('% of cycles')
ax.set_title('Seasonal distribution by cluster (%)')
ax.legend(title='Season')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 3) Year distribution (counts)
fig, ax = plt.subplots(figsize=(10, 4))
for i, (cluster, row) in enumerate(year_dist.iterrows()):
    ax.plot(row.index, row.values, marker='o', label=f'Cluster {cluster}')
ax.set_xlabel('Year')
ax.set_ylabel('Count of cycles')
ax.set_title('Cycles per year by cluster')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
