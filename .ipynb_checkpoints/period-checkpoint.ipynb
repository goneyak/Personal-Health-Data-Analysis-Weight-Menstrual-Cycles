### Step 1: Parse and print the XML structure

```python
import xml.etree.ElementTree as ET

# Load the XML file
tree = ET.parse("period.xml")
root = tree.getroot()

# Print root tag and its immediate children
print("Root tag:", root.tag)
print("Children:")
for child in root:
    print(" ", child.tag)
```

```python
# Explore menstruation_list and its first entry
menstruation_list = root.find("menstruation_list")
print("\nFirst few menstruation entries:")
for i, m in enumerate(menstruation_list.findall("menstruation")):
    print(f"\nEntry {i+1}:")
    for item in m:
        print(f"  {item.tag}: {item.text}")
    if i >= 2:  # only preview 3 entries
        break
```

```python
# Extract menstruation data
data = []
for menstruation in root.find("menstruation_list").findall("menstruation"):
    begin = menstruation.find("begin").text
    end = menstruation.find("end").text
    data.append({"begin": begin, "end": end})

# Create DataFrame
df_periods = pd.DataFrame(data)

# Convert date columns to datetime
df_periods["begin"] = pd.to_datetime(df_periods["begin"])
df_periods["end"] = pd.to_datetime(df_periods["end"])

# Calculate duration in days
df_periods["duration_days"] = (df_periods["end"] - df_periods["begin"]).dt.days + 1

# Sort by start date
df_periods = df_periods.sort_values("begin").reset_index(drop=True)

# Preview
print(df_periods.head())
```

```python
import matplotlib.pyplot as plt

plt.hist(df_periods['duration_days'], bins=range(1, 21))
plt.title('Distribution of Menstruation Duration')
plt.xlabel('Duration (days)')
plt.ylabel('Frequency')
plt.show()
```

### Step 2: Missing data detection and interpolation

```python
# Make sure data is sorted by begin date
df_periods = df_periods.sort_values("begin").reset_index(drop=True)

# Calculate days between consecutive period starts (cycle length)
df_periods["next_begin"] = df_periods["begin"].shift(-1)
df_periods["cycle_length"] = (df_periods["next_begin"] - df_periods["begin"]).dt.days

# Define threshold for missing period detection
threshold_days = 48

# Find indices where cycle length exceeds threshold
missing_periods = df_periods[df_periods["cycle_length"] > threshold_days]

print("Potential missing periods detected at these intervals:")
for idx, row in missing_periods.iterrows():
    expected_missing_start = row["begin"] + pd.Timedelta(days=28)
    expected_missing_end = row["next_begin"] - pd.Timedelta(days=1)
    print(f"- Missing period between {expected_missing_start.date()} and {expected_missing_end.date()} (gap of {row['cycle_length']} days)")
```

```python
import pandas as pd
import math

def fill_missing_periods_v4(df, threshold=48, default_duration=6):
    df = df.sort_values('begin').reset_index(drop=True)
    new_rows = []
    
    for i in range(len(df) - 1):
        current = df.loc[i, 'begin']
        next_ = df.loc[i + 1, 'begin']
        diff = (next_ - current).days
        
        if diff > threshold:
            prev_cycle = (current - df.loc[i - 1, 'begin']).days if i > 0 else 28
            next_cycle = (df.loc[i + 2, 'begin'] - next_).days if i + 2 < len(df) else 28
            avg_cycle = (prev_cycle + next_cycle) / 2
            
            n_missing = max(0, math.ceil(diff / avg_cycle) - 1)
            
            for j in range(n_missing):
                new_begin = current + pd.Timedelta(days=avg_cycle * (j + 1))
                # Assign default duration for imputed periods
                new_end = new_begin + pd.Timedelta(days=default_duration - 1)
                new_rows.append({
                    'begin': new_begin,
                    'end': new_end,
                    'duration_days': default_duration,
                    'imputed': True
                })
    
    df_new = pd.DataFrame(new_rows)
    df['imputed'] = False
    df_filled = pd.concat([df, df_new], ignore_index=True).sort_values('begin').reset_index(drop=True)
    return df_filled

df_filled = fill_missing_periods_v4(df_periods, threshold=48, default_duration=6)

print(df_filled[df_filled['imputed']].head())
print(f"Imputed count: {df_filled['imputed'].sum()}")
```

```python
# After filling missing periods
df_filled = fill_missing_periods_v5(df_periods, threshold=48, default_duration=6)

# Recalculate derived columns
df_filled['day'] = df_filled['begin'].dt.day
df_filled['start_day'] = df_filled['begin'].dt.day
df_filled['end_day'] = df_filled['end'].dt.day
df_filled['year_month'] = df_filled['begin'].dt.to_period('M').astype(str)

print(df_filled[df_filled['imputed']].head())
print(f"Imputed count: {df_filled['imputed'].sum()}")
```

### Step 3: Evaluate Ovulation dates

```python
# Sort and calculate cycle_length
df_filled = df_filled.sort_values('begin').reset_index(drop=True)
df_filled['next_begin'] = df_filled['begin'].shift(-1)
df_filled['cycle_length'] = (df_filled['next_begin'] - df_filled['begin']).dt.days

# Calculate ovulation_date
df_filled['ovulation_date'] = df_filled.apply(
    lambda row: row['begin'] + pd.Timedelta(days=row['cycle_length'] - 14)
    if pd.notnull(row['cycle_length']) else pd.NaT,
    axis=1
)

print(df_filled[['begin', 'cycle_length', 'ovulation_date']].head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.tseries.offsets import MonthEnd

# (기존 코드)
df_periods['start'] = df_periods['begin']
df_periods['end'] = df_periods['end']

split_rows = []

for idx, row in df_periods.iterrows():
    start = row['start']
    end = row['end']
    
    period_months = pd.period_range(start=start.to_period('M'), end=end.to_period('M'), freq='M')
    
    for month_period in period_months:
        year_month = str(month_period)
        month_start = month_period.to_timestamp()
        month_end = month_start + MonthEnd(0)
        
        segment_start = max(start, month_start)
        segment_end = min(end, month_end)
        
        start_day = segment_start.day
        end_day = segment_end.day
        
        split_rows.append({
            'year_month': year_month,
            'start_day': start_day,
            'end_day': end_day
        })

df_split = pd.DataFrame(split_rows)

labels = sorted(df_split['year_month'].unique())
label_to_y = {label: i for i, label in enumerate(labels)}

plt.figure(figsize=(15, len(labels)*0.3))

# Draw menstruation bars
for _, row in df_split.iterrows():
    y = label_to_y[row['year_month']]
    width = row['end_day'] - row['start_day'] + 1
    plt.barh(
        y=y,
        width=width,
        left=row['start_day'],
        height=0.6,
        color='salmon',
        edgecolor='black'
    )

# --- 배란일 표시 추가 시작 ---

# ovulation_date 정보가 포함된 DataFrame (예: df_filled)
# ovulation_date가 NaT 아닌 행만 사용
df_ovulation = df_filled.dropna(subset=['ovulation_date']).copy()

# ovulation 연-월, 일 추출
df_ovulation['year_month'] = df_ovulation['ovulation_date'].dt.to_period('M').astype(str)
df_ovulation['ovulation_day'] = df_ovulation['ovulation_date'].dt.day

# ovulation 점 그리기
for _, row in df_ovulation.iterrows():
    y = label_to_y.get(row['year_month'])
    if y is not None:
        plt.scatter(
            row['ovulation_day'], y, 
            color='blue', s=30, marker='D', label='Ovulation Day' if 'Ovulation Day' not in plt.gca().get_legend_handles_labels()[1] else ""
        )

# --- 배란일 표시 추가 끝 ---

plt.yticks(range(len(labels)), labels)
plt.xlabel('Day of Month')
plt.title('Menstruation Period Duration and Ovulation Days by Year-Month')

plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
