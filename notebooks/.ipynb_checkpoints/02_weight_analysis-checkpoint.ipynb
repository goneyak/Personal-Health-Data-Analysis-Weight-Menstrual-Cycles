### Step 1: Load the data and select only the necessary columns

```python
import pandas as pd

# Weight data loading
df_weight = pd.read_excel("weight.xlsx")

# Check data information
print(df_weight.info())
print("Number of missing data:", df_weight.isnull().sum())

# Top 5 rows
print("weight.xlsx:")
display(df_weight.head())
```

```python
# Rename columns for easier access (optional)
df_weight = df_weight.rename(columns={'Date [Europe/Berlin]': 'Date', 'Weight [kg]': 'Weight'})

# Select only Date and Weight columns
df_weight = df_weight[['Date', 'Weight']]

# Convert the Date column to datetime
df_weight['Date'] = pd.to_datetime(df_weight['Date'])

# Sort the data by date
df_weight = df_weight.sort_values('Date').reset_index(drop=True)

# Display the first few rows
print(df_weight.head())
```

### Step 2: Check for missing dates between first and last measurements

```python
# Get the start and end date from the data
start_date = df_weight['Date'].min().normalize()
end_date = df_weight['Date'].max().normalize()

# Create a complete date range between the start and end date
full_date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Extract just the date part from the Date column (drop time)
df_weight['OnlyDate'] = df_weight['Date'].dt.date

# Convert to datetime for comparison
existing_dates = pd.to_datetime(df_weight['OnlyDate'].unique())

# Find missing dates by comparing to full range
missing_dates = full_date_range.difference(existing_dates)

# Display result
print(f"Total missing days: {len(missing_dates)}")
print(missing_dates)
```

### Step 3: Visualize missing dates

```python
import matplotlib.pyplot as plt

# Plot settings
plt.figure(figsize=(15, 2))

# Plot all days in the date range as grey dots (background)
plt.plot(full_date_range, [1] * len(full_date_range), '|', color='lightgray', markersize=12, label='All Dates')

# Plot existing dates in blue
plt.plot(existing_dates, [1] * len(existing_dates), '|', color='blue', markersize=12, label='Measured Dates')

# Plot missing dates in red
plt.plot(missing_dates, [1] * len(missing_dates), '|', color='red', markersize=12, label='Missing Dates')

# Formatting
plt.yticks([])
plt.xlabel('Date')
plt.title('Missing vs. Measured Dates')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

### Step 4: Analyze data density by quarter

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract date-only and quarter info
df_weight['OnlyDate'] = df_weight['Date'].dt.date
df_weight['Quarter'] = df_weight['Date'].dt.to_period('Q')

# Count unique measurement days per quarter
measured_days_per_quarter = df_weight.drop_duplicates('OnlyDate').groupby('Quarter').size()

# Create full date range from start to end date
start_date = df_weight['Date'].min().normalize()
end_date = df_weight['Date'].max().normalize()
full_date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Map each day in the full range to its quarter
full_quarters = pd.Series(full_date_range).dt.to_period('Q')
total_days_per_quarter = full_quarters.value_counts().sort_index()

# Calculate density: recorded days / total days per quarter
density = (measured_days_per_quarter / total_days_per_quarter).fillna(0)

# Plot the density
plt.figure(figsize=(14, 5))
density.plot(kind='bar', color='lightgreen')
plt.title('Measurement Density by Quarter')
plt.ylabel('Recorded Days / Total Days')
plt.xlabel('Quarter')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 5: Filter data from 2021 Q3 onward

```python
# Make sure 'Date' column is in datetime format
df_weight['Date'] = pd.to_datetime(df_weight['Date'])

# Filter from 2021 Q3 onward
df_filtered = df_weight[df_weight['Date'] >= '2021-07-01'].copy()
```

### Step 6: Compute daily average weight (one row per day)

```python
# Check for duplicate dates (after removing time)
df_temp = df_filtered.copy()
df_temp['OnlyDate'] = df_temp['Date'].dt.normalize()

# Count how many times each date appears
date_counts = df_temp['OnlyDate'].value_counts()

# Filter dates that appear more than once
duplicate_dates = date_counts[date_counts > 1]

# Show the result
print("Number of days with multiple measurements:", len(duplicate_dates))
print(duplicate_dates.sort_index())
```

```python
# Compute daily average weight
df_daily_avg = df_filtered.copy()

# Normalize the datetime to just the date (remove time)
df_daily_avg['OnlyDate'] = df_daily_avg['Date'].dt.normalize()

# Group by date and calculate average weight
df_daily_avg = df_daily_avg.groupby('OnlyDate')['Weight'].mean().reset_index()

# Rename the date column for consistency
df_daily_avg = df_daily_avg.rename(columns={'OnlyDate': 'Date'})

# Preview the result
print(df_daily_avg.head())
```

### Step 7: Add missing dates (fill gaps with NaN rows)

```python
# Create a full date range and reindex to fill missing days
start_date = df_daily_avg['Date'].min()
end_date = df_daily_avg['Date'].max()

# Generate a complete daily date range
full_dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Set 'Date' as the index for reindexing
df_daily_filled = df_daily_avg.set_index('Date').reindex(full_dates)

# Reset index and rename it back to 'Date'
df_daily_filled.index.name = 'Date'
df_daily_filled = df_daily_filled.reset_index()
```

```python
# Plot weight over time (NaNs will appear as gaps)
plt.figure(figsize=(14, 6))
plt.plot(df_daily_filled['Date'], df_daily_filled['Weight'], label='Weight', color='blue')

plt.title('Daily Weight Over Time')
plt.xlabel('Date')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 8: Linear Interpolation for Missing data

```python
# Copy to avoid modifying the original
df_interpolated = df_daily_filled.copy()

# Sort by date to ensure interpolation works properly
df_interpolated = df_interpolated.sort_values('Date')

# Linear interpolation
df_interpolated['Weight'] = df_interpolated['Weight'].interpolate(method='linear')

# Fill any remaining NaNs at the start/end using nearest values
df_interpolated['Weight'] = df_interpolated['Weight'].bfill().ffill()

# Preview result
print(df_interpolated.head())
print("Remaining missing values:", df_interpolated['Weight'].isna().sum())
```

```python
# Compute summary stats
average_weight = df_interpolated['Weight'].mean()
max_weight = df_interpolated['Weight'].max()
min_weight = df_interpolated['Weight'].min()

# Find corresponding dates
max_date = df_interpolated.loc[df_interpolated['Weight'].idxmax(), 'Date']
min_date = df_interpolated.loc[df_interpolated['Weight'].idxmin(), 'Date']

# Plot
plt.figure(figsize=(14, 6))

# Interpolated line
plt.plot(df_interpolated['Date'], df_interpolated['Weight'], label='Weight (Interpolated)', color='orange')

# Average line
plt.axhline(y=average_weight, color='green', linestyle='--', linewidth=2, label=f'Average ({average_weight:.2f} kg)')

# Max point
plt.scatter(max_date, max_weight, color='red', label=f'Max ({max_weight:.2f} kg)', zorder=5)
plt.text(max_date, max_weight + 0.2, f'{max_weight:.2f} kg', color='red', ha='center')

# Min point
plt.scatter(min_date, min_weight, color='blue', label=f'Min ({min_weight:.2f} kg)', zorder=5)
plt.text(min_date, min_weight - 0.4, f'{min_weight:.2f} kg', color='blue', ha='center')

# Labels and grid
plt.title('Daily Weight Over Time (After Interpolation)')
plt.xlabel('Date')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Copy data to avoid modifying original
df_ma = df_interpolated.copy()

# Compute moving averages
df_ma['MA_7'] = df_ma['Weight'].rolling(window=7, min_periods=1).mean()
df_ma['MA_30'] = df_ma['Weight'].rolling(window=30, min_periods=1).mean()

# Plot
plt.figure(figsize=(16, 6))
plt.plot(df_ma['Date'], df_ma['Weight'], label='Daily Weight', color='lightgray', alpha=0.5)
plt.plot(df_ma['Date'], df_ma['MA_7'], label='7-Day MA', color='orange')
plt.plot(df_ma['Date'], df_ma['MA_30'], label='30-Day MA', color='blue')
plt.title('Daily Weight with 7-Day and 30-Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 9: Caclulate Montly Average

```python
# Add 'Month' column
df_month = df_interpolated.copy()
df_month['Month'] = df_month['Date'].dt.month

# Group by month
month_avg = df_month.groupby('Month')['Weight'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(month_avg.index, month_avg.values, marker='o', color='darkgreen')
plt.title('Average Weight by Month')
plt.xlabel('Month')
plt.ylabel('Average Weight (kg)')
plt.xticks(range(1, 13))
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
mean_weight = month_avg.mean()
delta_month = month_avg - mean_weight

plt.figure(figsize=(8, 4))
plt.bar(delta_month.index, delta_month.values, color='steelblue')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Deviation from Overall Average Weight by Month')
plt.xlabel('Month')
plt.ylabel('Δ Weight (kg)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

groups = [group['Weight'].values for _, group in df_month.groupby('Month')]
f_stat, p_val = stats.f_oneway(*groups)
print(f"ANOVA p-value (Difference in Month): {p_val}")
```

```python
# Montly Average Weight by Year
df_monthly = df_interpolated.copy()
df_monthly['Year'] = df_monthly['Date'].dt.year
df_monthly['Month'] = df_monthly['Date'].dt.month

# Monthly Average Weight
df_grouped = df_monthly.groupby(['Year', 'Month'])['Weight'].mean().reset_index()

plt.figure(figsize=(12, 6))

for year in df_grouped['Year'].unique():
    data = df_grouped[df_grouped['Year'] == year]
    plt.plot(data['Month'], data['Weight'], marker='o', label=str(year))

plt.title('Monthly Average Weight by Year')
plt.xlabel('Month')
plt.ylabel('Average Weight (kg)')
plt.xticks(range(1, 13))
plt.legend(title='Year')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Step 10: Season-to-Season Weight Changes by Year

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df_seasonal = df_interpolated.copy()
df_seasonal['Year'] = df_seasonal['Date'].dt.year
df_seasonal['Month'] = df_seasonal['Date'].dt.month
df_seasonal['Season'] = df_seasonal['Month'].apply(get_season)

seasonal_avg = df_seasonal.groupby(['Year', 'Season'])['Weight'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=seasonal_avg, x='Season', y='Weight', hue='Year', marker='o')
plt.title('Seasonal Average Weight by Year')
plt.ylabel('Average Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
pivot = seasonal_avg.pivot(index='Season', columns='Year', values='Weight')
pivot = pivot.reindex(['Winter', 'Spring', 'Summer', 'Fall'])

plt.figure(figsize=(8, 4))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Seasonal Average Weight Heatmap")
plt.xlabel("Year")
plt.ylabel("Season")
plt.tight_layout()
plt.show()
```

```python
# Pivot to create season columns per year
season_table = seasonal_avg.pivot(index='Year', columns='Season', values='Weight')
season_table = season_table[['Winter', 'Spring', 'Summer', 'Fall']]  # ensure order

# Compute within-year deltas
delta_table = pd.DataFrame(index=season_table.index)
delta_table['Δ Spring - Winter'] = season_table['Spring'] - season_table['Winter']
delta_table['Δ Summer - Spring'] = season_table['Summer'] - season_table['Spring']
delta_table['Δ Fall - Summer'] = season_table['Fall'] - season_table['Summer']

# Compute cross-year delta (Fall → next Winter)
delta_next_winter = []
for year in season_table.index:
    next_year = year + 1
    if next_year in season_table.index:
        fall = season_table.loc[year, 'Fall']
        next_winter = season_table.loc[next_year, 'Winter']
        if pd.notna(fall) and pd.notna(next_winter):
            delta = next_winter - fall
        else:
            delta = None
    else:
        delta = None
    delta_next_winter.append(delta)

# Add Fall → Next Winter delta to table
delta_table['Δ Next Winter - Fall'] = delta_next_winter

# Round values and display
delta_table = delta_table.round(2)

print("Yearly Season-to-Season Weight Change Table:")
display(delta_table)
```

```python
# Define the x-axis labels (season transitions)
season_steps = ['Winter→Spring', 'Spring→Summer', 'Summer→Fall', 'Fall→Next Winter']

# Plotting setup
plt.figure(figsize=(10, 4))

# Iterate over each year
for year, row in delta_table.iterrows():
    # Get the 4 delta values in order
    y_values = [
        row['Δ Spring - Winter'],
        row['Δ Summer - Spring'],
        row['Δ Fall - Summer'],
        row['Δ Next Winter - Fall']
    ]
    
    # Skip if all values are NaN
    if pd.isnull(y_values).all():
        continue

    # Plot the line for the year
    plt.plot(season_steps, y_values, marker='o', label=str(year), alpha=0.8)

# Labels and styling
plt.title('Seasonal Weight Changes by Year')
plt.xlabel('Season Transition')
plt.ylabel('Δ Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Year')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.tight_layout()
plt.show()
```
